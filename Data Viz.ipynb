```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].map({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})

# Task 1: Data Exploration
print(data.head())
print(data.info())
print(data.describe())

# Task 2: Basic Data Analysis
grouped = data.groupby('species').mean()
print(grouped)

# Task 3: Visualizations
# Line Chart
plt.plot(data['sepal length (cm)'], label="Sepal Length")
plt.title("Line Chart: Sepal Length")
plt.xlabel("Index")
plt.ylabel("Length (cm)")
plt.legend()
plt.show()

# Bar Chart
sns.barplot(x="species", y="sepal width (cm)", data=data)
plt.title("Bar Chart: Sepal Width by Species")
plt.show()

# Histogram
plt.hist(data['petal length (cm)'], bins=20, color="orange")
plt.title("Histogram: Petal Length Distribution")
plt.show()

# Scatter Plot
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="species", data=data)
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
```
